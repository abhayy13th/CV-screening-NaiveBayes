# Data Cleaning

```python
resume_df ['resume_text'] = resume_df ['resume_text'].apply(lambda x:x.replace('\r',''))
resume_df 
```

```python
nltk.download('punkt')
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from','subject','re','edu','com','email','use'])
```

```python
# Remove stop words and remove words with 2 or less characters
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 2 and token not in stop_words:
            result.append(token)
            
    return ' '.join(result)
```

```python
#text after cleaning
resume_df['cleaned'] = resume_df['resume_text'].apply(preprocess)
resume_df
```

<h3>Prepare data set for training </h3>


```python
#count vectorization

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(resume_df['cleaned'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(X.toarray())
```

```python
y=resume_df['class']
```

```python
X.shape
```

```python
y.shape
```

<h3> Data Splitting </h3>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_predict_train = NB_classifier.predict(X_train)
y_predict_train
c = confusion_matrix(y_train,y_predict_train)
sns.heatmap(c,annot=True)
```

<h3> Results</h3>

```python
y_predict_test = NB_classifier.predict(X_test)

c = confusion_matrix(y_test,y_predict_test)
sns.heatmap(c,annot=True)
```

```python
print(classification_report(y_test,y_predict_test))
```
